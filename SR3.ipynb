```python
import numpy as np
import seaborn as sns
data = np.random.normal(loc=10.0, scale=5.0, size=8130)
plt.figure(figsize=(12, 6))
sns.kdeplot(data, shade=True, color='purple', label='Density Plot')
sns.rugplot(data, color='black')
plt.title('Density Plot with Rug Plot')
plt.xlabel('Value')
plt.ylabel('Country Name')
plt.grid(True)
plt.legend()
plt.show()
```
